```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LogisticRegression
import pandas as pd
from sklearn import metrics
from matplotlib import pyplot as plt

data = pd.read_csv("https://raw.githubusercontent.com/koroteevmv/ML_course/2023/ML2.2%20real%20classification/data/diabetes.csv")
X = data.drop('Outcome', axis = 1)
Y = data['Outcome']

logreg = LogisticRegression()
logreg.fit(X, Y)
pred = logreg.predict(X)

from sklearn import metrics
print(metrics.confusion_matrix(Y, pred))
print(logreg.score(X, Y))
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree = 2)
X_poly = poly.fit_transform(X)

poly_logreg = LogisticRegression()
poly_logreg.fit(X_poly, Y)
pred = poly_logreg.predict(X_poly)

print(metrics.confusion_matrix(Y, pred))
print(poly_logreg.score(X_poly, Y))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

def different_models(X, Y):
    models = {
        'LinearRegression': Pipeline([('scaler', StandardScaler()), ('model', LogisticRegression())]),
        'SVR': Pipeline([('scaler', StandardScaler()), ('model', SVC())]),
        'SVR_lin': Pipeline([('scaler', StandardScaler()), ('model', SVC(kernel='linear'))]),
        'SVR_rbf': Pipeline([('scaler', StandardScaler()), ('model', SVC(kernel='rbf', C=10))]),
        'SVR_pol': Pipeline([('scaler', StandardScaler()), ('model', SVC(kernel='poly', degree=3))]),
        'KNN': Pipeline([('scaler', StandardScaler()), ('model', KNeighborsClassifier(n_neighbors=5))]),
        'MLP': Pipeline([('scaler', StandardScaler()), ('model', MLPClassifier(max_iter=3000))]),
        'DecisionTree': DecisionTreeClassifier(max_depth=5),
        'NaiveBayes': GaussianNB()
    }

    for name, model in models.items():
        model.fit(X, Y)
        predicted = model.predict(X)
        print(f'{name} R2', model.score(X, Y))
        print(metrics.confusion_matrix(Y, predicted))
        print(model.score(X, Y))
different_models(X, Y)
```
