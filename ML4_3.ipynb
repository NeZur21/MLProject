```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
import numpy as np

iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)

X = data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=5)

model1 = LogisticRegression()
model1.fit(X_train, y_train)
y_pred = model1.predict(X_test)
print(accuracy_score(y_test, y_pred))
print(f1_score(y_test, y_pred, average='macro'))
print(confusion_matrix(y_test, y_pred))
print()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=0)

model2 = LogisticRegression()
model2.fit(X_train, y_train)
y_pred = model2.predict(X_test)
print(accuracy_score(y_test, y_pred))
print(f1_score(y_test, y_pred, average='macro'))
print(confusion_matrix(y_test, y_pred))
print()


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=8)

model3 = LogisticRegression()
model3.fit(X_train, y_train)
y_pred = model3.predict(X_test)
print(accuracy_score(y_test, y_pred))
print(f1_score(y_test, y_pred, average='macro'))
print(confusion_matrix(y_test, y_pred))
print()
```

```python
from sklearn.model_selection import KFold,StratifiedKFold,LeaveOneOut, cross_val_score

X_np = X.values

kf = KFold(n_splits=3, shuffle=True, random_state=15)
for i, (train_index, test_index) in enumerate(kf.split(y)):
    print(f'Fold {i+1}: Длина train: {len(train_index)}, test: {len(test_index)}')
    print(f'Train: {train_index}, Test: {test_index}')
    print()

metrics_accuracy = []
metrics_f1 = []
model = LogisticRegression(max_iter=1000)
for i, (train_index, test_index) in enumerate(kf.split(y)):
    X_train, X_test = X_np[train_index], X_np[test_index]
    y_train, y_test = y[train_index], y[test_index]
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    metrics_accuracy.append(accuracy_score(y_test, y_pred))
    metrics_f1.append(f1_score(y_test, y_pred, average='macro'))

print(f'Значения метрик accuracy: {metrics_accuracy} \n Значение метрик f1: {metrics_f1}')

print(f'Среднее по кросс-валидации: {np.array(metrics_accuracy).mean()}')
print(f'Среднее по кросс-валидации: {np.array(metrics_f1).mean()}')

```

```python
cv_result = cross_val_score(model, X_np, y, cv=kf, scoring='accuracy', n_jobs=-1)

print('Кросс-валидация: ', cv_result)
print('Среднее по кросс-валидация: ', cv_result.mean())
print('Дисперсия по кросс-валидации: ', cv_result.std())
print()

loo = LeaveOneOut()

for i, (train_index, test_index) in enumerate(loo.split(y)):
    print(f'Fold {i+1}: Длина train: {len(train_index)}, test: {len(test_index)}')
    print(f'Train: {train_index}, Test: {test_index}')
    print()

cv_result = cross_val_score(model, X_np, y, cv=loo, scoring='f1_macro', n_jobs=-1)

print('Кросс-валидация: ', cv_result)
print('Среднее по кросс-валидация: ', cv_result.mean())
print('Дисперсия по кросс-валидации: ', cv_result.std())
print()
```
